**Import Libraries**

```python
import numpy as np
import matplotlib.pyplot as plt
```

**Get User Input for Triangle**

```python
def get_triangle():
    print("Enter the 3 vertices of the triangle:")
    points = []
    for i in range(3):
        x = float(input(f"  Point {i+1} - x: "))
        y = float(input(f"  Point {i+1} - y: "))
        points.append((x, y))
    points.append(points[0])  # Close the triangle
    return np.array(points)
```

**Plotting Function**

```python
def plot_transformation(original, new_points, title):
    plt.figure(figsize=(5, 4))
    plt.plot(*original.T, label='Original', linewidth=2)
    plt.plot(*new_points.T, label='Transformed')
    plt.title(title)
    plt.legend()
    plt.grid(True)
    plt.axis('equal')
    plt.tight_layout()
    plt.show()
```

**Translate**

```python
def translate(points, dx, dy):
    return np.array([(x + dx, y + dy) for x, y in points])
```

**Scaling**

```python
def scale(points, sx, sy):
    return np.array([(x * sx, y * sy) for x, y in points])
```

**Rotating**

```python
def rotate(points, angle_deg):
    theta = np.radians(angle_deg)
    R = np.array([[np.cos(theta), -np.sin(theta)],
                  [np.sin(theta),  np.cos(theta)]])
    return np.dot(points, R.T)
```

**Refection**

```python
def reflect(points, axis='x'):
    matrices = {
        'x': np.array([[1, 0], [0, -1]]),
        'y': np.array([[-1, 0], [0, 1]]),
        'origin': np.array([[-1, 0], [0, -1]]),
        'y=x': np.array([[0, 1], [1, 0]])
    }
    return np.dot(points, matrices[axis].T)
```

**Shearing**

```python
def shear(points, shx=0, shy=0):
    shear_matrix = np.array([[1, shx], [shy, 1]])
    return np.dot(points, shear_matrix.T)
```

**Run Program**

```python
triangle = get_triangle()
```

**Apply and Plot Each Transfomations**

**Translation Plotting:**

```python
plot_transformation(triangle, translate(triangle, 3, 2), "Translation (3, 2)")
```

**Scaling Plotting:**

```python
plot_transformation(triangle, scale(triangle, 1.5, 0.5), "Scaling (1.5x, 0.5y)")
```

**Rotation Plotting:**

```python
plot_transformation(triangle, rotate(triangle, 45), "Rotation (45Â° CCW)")
```

**Reflection Plotting:**

```python
plot_transformation(triangle, reflect(triangle, 'x'), "Reflection over X-axis")
plot_transformation(triangle, reflect(triangle, 'y'), "Reflection over Y-axis")
```

Shearing

```python
plot_transformation(triangle, shear(triangle, shx=2), "Shearing X-axis(shy=2)")
plot_transformation(triangle, shear(triangle, shy=2), "Shearing Y-axis (shy=2)")
```
